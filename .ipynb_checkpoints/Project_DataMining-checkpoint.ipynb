```python
RDDread = sc.textFile("crime_homicide_subset.csv") \
            .filter(lambda x : x.split(',')[0] == '2012') \
            .map(lambda y : (int(y.split(',')[1]), int(1))) \
            .reduceByKey(lambda x, y : x + y) \
            .sortByKey(ascending=True)

first = sc.parallelize([(0, 0)])

goodRDD = first.union(RDDread)

goodRDD.collect()
            
lines = goodRDD.values().collect()


fig = plt.figure(figsize=(15, 5))
fig.suptitle('Number of crimes per month for 2012', fontsize=14, fontweight='bold')

x_ticks = np.arange(goodRDD.keys().min(), goodRDD.keys().max() + 1, 1)
y_ticks = np.arange(goodRDD.values().min(), goodRDD.values().max(), 5)

number = fig.add_subplot(122)

number.plot(lines)
number.set_xlabel("Month")
number.set_ylabel("Number of crimes")
number.set_xticks(x_ticks)
number.set_yticks(y_ticks)
number.spines['top'].set_color('none')
number.spines['right'].set_color('none')
number.xaxis.set_ticks_position('bottom')
number.yaxis.set_ticks_position('left')


plt.show()
```

```python
import matplotlib.pyplot as plt

header = sc.textFile("race.csv").take(1)[0]

data = sc.textFile("race.csv").filter(lambda l : l != header).map(lambda x : (x.split(',')[0], x.split(',')[1])).keys().distinct().sortBy(lambda y : y[0])


wards = tuple(data.keys().collect())
        

w = widgets.Dropdown(
    options=wards,
    description='Ward:',
)

RDDread = sc.textFile("race.csv")
header = RDDread.take(1)[0]
data = RDDread.filter(lambda l: l != header)


cols = ['#9FB5F3', '#ECA9B4', '#A4F39F', '#F3A9ED', '#A740EC', '#99ECE7', '#EBEB4A', '#F2911A']

def change_pie(value):
    fig = plt.figure(figsize=(15, 10))
    fig.suptitle("Ward " + value, fontsize=14, fontweight='bold')
    pieke = fig.add_subplot(121)
    data2 = data.filter(lambda l: l.startswith(value))
    data2 = data2.map(lambda x: (x.split(',')[1], float(x.split(',')[3])))
    slices = data2.values().collect()
    ethnicgroups = data2.keys().collect()
    i=0
    for x in ethnicgroups:
        s = ''
        for y in x.split('_'):
            s = s + y + ' '
        ethnicgroups[i] = s.strip()
        i = i+1
    pieke.pie(slices, labels=ethnicgroups, colors=cols, autopct='%1.1f%%')
    plt.draw()
    
def change_handler(change):
    change_pie(change['new'])
      

w.observe(change_handler, names='value')


display(w)
change_pie(w.value)
```
